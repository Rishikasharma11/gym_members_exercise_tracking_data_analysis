```python
data.columns = data.columns.str.replace('_', ' ')
data
```

```python
high_calorie_burners = data[data['Calories Burned'] > 500]
high_calorie_burners
```

```python
def bmi_category(bmi):
    if bmi < 18.5:
        return 'Underweight'
    elif 18.5 <= bmi < 24.9:
        return 'Normal weight'
    elif 25 <= bmi < 29.9:
        return 'Overweight'
    else:
        return 'Obese'

data['BMI_Category'] = data['BMI'].apply(bmi_category)
data[['BMI', 'BMI_Category']]
```

```python
group_by_gender_workout = data.groupby(['Gender', 'Workout Type'])['Calories Burned'].mean()
group_by_gender_workout
```

```python
pivot_table = data.pivot_table(values='Fat Percentage', index='Gender', columns='Workout Type', aggfunc='mean')
pivot_table
```

```python
sorted_data = data.sort_values(by='Calories Burned', ascending=False)
sorted_data[['Calories Burned', 'Gender', 'Workout Type']].head(10) 
```

```python
import plotly.express as px

fig = px.scatter_matrix(data.head(50), dimensions=['Weight (kg)', 'Calories Burned', 'BMI'],
                        color='Gender', hover_data=['Age'], title='Scatter Matrix: Weight, Calories Burned, and BMI')
fig.show()
```

```python
fig = px.scatter(data.head(50), x='Weight (kg)', y='Calories Burned', color='Gender',
                 title='Weight vs Calories Burned', size='BMI', hover_data=['Age'])
fig.show()
```
